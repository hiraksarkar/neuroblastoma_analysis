```r
store_dir = "/scratch/gpfs/KANG/hirak/mendel/Projects/neuroblastoma_analysis/results/velocity/5_sample/"
```

```r


for (name in names(hemi.obj)) {


  output_dir <- glue::glue(store_dir, name)
  if (!dir.exists(output_dir)) {
      dir.create(output_dir, recursive = TRUE)
  }
  # Extract the raw count matrix
  sparse_matrix <- hemi.obj[[name]]@assays$RNA$counts
  
  # Convert the matrix to a sparse matrix if it isn't already
  # sparse_matrix <- as(matrix, "CsparseMatrix")
  
  # Define the file paths
  mtx_file <- file.path(output_dir, paste0(name, "_matrix.mtx"))
  barcodes_file <- file.path(output_dir, paste0(name, "_barcodes.tsv"))
  features_file <- file.path(output_dir, paste0(name, "_features.tsv"))
  umap_file <- file.path(output_dir, paste0(name, "_umap.csv"))
  metadata_file <- file.path(output_dir, paste0(name, "_metadata.csv"))
  
  # Save the matrix as an MTX file
  Matrix::writeMM(sparse_matrix, file = mtx_file)
  
  # Save the barcodes (cell names)
  write.table(rownames(sparse_matrix), file = barcodes_file, 
              quote = FALSE, sep = "\t", col.names = FALSE, row.names = FALSE)
  
  # Save the features (gene names)
  write.table(colnames(sparse_matrix), file = features_file, 
              quote = FALSE, sep = "\t", col.names = FALSE, row.names = FALSE)
  
  # Extract and save UMAP embeddings
  umap_df <- Embeddings(hemi.obj[[name]], reduction = "umap")
  write.csv(umap_df, file = umap_file, row.names = TRUE)
  
  # Extract and save the entire metadata
  metadata_df <- hemi.obj[[name]]@meta.data
  write.csv(metadata_df, file = metadata_file, row.names = TRUE)
}
```
