```r
# Initialize a list to hold Seurat objects
seurat_list <- list()

# Loop over each file, read it, and create a Seurat object
for (file in file_paths) {
    # Extract the sample name from the file name
    sample_name <- gsub(".dge.txt.gz", "", file)
    
    # Get the corresponding library ID
    library_id <- library_ids[sample_name]
    
    # Read the data
    data <- read.table(
        paste0('/scratch/gpfs/KANG/hirak/mendel/Projects/neuroblastoma_analysis/results/revision/',file), 
        header = TRUE, 
        row.names = 1)
    
    # Create a Seurat object and assign the library ID as the project name
    seurat_obj <- CreateSeuratObject(counts = data, project = library_id, min.cells = 3, min.features = 200)
    
    # Add the Seurat object to the list
    seurat_list[[library_id]] <- seurat_obj
}
```
